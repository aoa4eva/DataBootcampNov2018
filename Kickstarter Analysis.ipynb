```python
import pandas as pd
import numpy as np
location = 'Kickstartercombo.csv'
df = pd.read_csv(location)
df.head()
```

```python
#currency count overall
df['currency'].unique()
```

```python
#removed extra column 2nd index
df.drop('Unnamed: 0', axis=1)
```

```python

df['goal'].std()
```

```python
#mean for goal
df['goal'].mean()
```

```python
#mean for pledged
df['pledged'].mean()

```

```python
#most common outcome for kickstarters
df['state'].mode()
```

```python
#count total for grouping by state of fundraiser
states = df['state'].groupby(df['state']).count()
```

```python
states.head()
```

```python
#copy of states to show counts of fundraisers by main category
copy_states = df['state'].groupby(df['main_category']).count()
```

```python
copy_states
```

```python
#Pivot table describing avg backer, goal, pledged $ for each fundraiser by state and main category
pd.pivot_table(df, index=['state','main_category'],values=['goal','pledged','backers'])
```

```python
#Pivot table describing avg backer, goal, pledged $ for each fundraiser by main category
pd.pivot_table(df, index=['main_category'],values=['goal','pledged','backers'])
```

```python
#Pivot table describing avg backer, goal, pledged $ for each fundraiser by state, main category and category
pd.pivot_table(df, index=['state','main_category','category'],values=['goal','pledged','backers'])
```

```python
#Pivot table describing avg backer, goal, pledged $ for each fundraiser by country, currency and state
pd.pivot_table(df, index=['country', 'currency', 'state'],values=['goal','pledged','backers'])
```
